```python
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.shape
```

```python
cars.head()
```

# Strip Plot

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4, 6, 8])]
```

```python
sns.stripplot(x=cars.weight);
```

```python
sns.stripplot(x=cars.weight, y=cars.origin);
```

```python
sns.stripplot(x='weight', y='origin', hue='cylinders',
              data=cars
              );
```

```python
sns.stripplot(x='weight', y='origin', hue='cylinders',
              data=cars,
              dodge=True
              );
```

```python
sns.stripplot(x='origin', y='weight',
              data=cars
              );
```

# Styling

```python
sns.stripplot(x=cars.weight, y=cars.origin, jitter=0.05);
```

```python
from ipywidgets import interactive

def jitter_widget(jitter=0.1):
    sns.stripplot(x=cars.weight, y=cars.origin, jitter=jitter)
```

```python
interactive(jitter_widget, jitter=(0, 0.65, 0.05))
```

```python
sns.set_style('whitegrid')
```

```python
sns.stripplot(x=cars.weight, y=cars.origin, alpha=0.5);
```

```python
sns.stripplot(x=cars.weight, y=cars.origin,
              alpha=0.5,
              size=5
              );
```

```python
sns.stripplot(x=cars.weight, y=cars.origin, linewidth=1);
```

```python
sns.stripplot(x=cars.weight, y=cars.origin, linewidth=1, edgecolor='black');
```

```python
sns.stripplot(x=cars.weight, y=cars.origin, linewidth=1,
              alpha=0.7, size=10, jitter=0.25, marker='*'
              );
```
