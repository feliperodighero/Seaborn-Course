```python
import seaborn as sns
```

```python
sns.__version__
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
penguins.dropna(inplace=True)
```

```python
penguins.shape
```

# Histplot

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(x='bill_length_mm', data=penguins);
```

```python
sns.histplot(y='bill_length_mm', data=penguins);
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             kde=True
             );
```

```python
sns.histplot(x='bill_length_mm', data=penguins, bins=20);
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             bins=[30, 40, 45, 50, 60]
             );
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             binwidth=5, binrange=(30, 60)
             );
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             stat='count'
             );
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             stat='density'
             );
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             stat='probability'
             );
```

## cumulative

```python
sns.histplot(x='bill_length_mm', data=penguins,
             stat='probability', fill=False, element='step',
             cumulative=True
             );
```

# Categorical Variables

```python
sns.histplot(x='bill_length_mm', data=penguins,
             hue='species'
            );
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             hue='species',
             element='step'
            );
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             hue='species',
             element='poly'
            );
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             hue='species',
             multiple='layer'
            );
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             hue='species',
             multiple='stack'
            );
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             hue='species',
             multiple='fill'
            );
```

# Bivariate Histplot

```python
sns.set_style('white')
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins,
             cbar=True
            );
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins,
             hue='species'
            );
```

```python
sns.histplot(x='bill_length_mm', y='species', data=penguins,
             hue='species'
            );
```

# Styling

```python
sns.histplot(x='species', hue='sex', multiple='dodge', data=penguins,
             shrink=0.8
            );
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             color='indigo'
            );
```

```python
sns.histplot(x='species', hue='sex', multiple='dodge', data=penguins,
             shrink=0.8, palette='bone'
            );
```

```python
sns.histplot(x='bill_length_mm', data=penguins,
             fill=False
             );
```
