```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.shape
```

```python
diamonds.head()
```

# Scatter Plot

```python
sns.set_style('dark')
```

```python
diamonds = diamonds[
    diamonds.cut.isin(['Premium', 'Good']) &
    diamonds.color.isin(['D', 'F', 'J'])
].sample(n=100, random_state=42)

diamonds.shape
```

```python
sns.scatterplot(x=diamonds.carat, y=diamonds.price);
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds);
```

```python
sns.scatterplot(x='x', y='y', data=diamonds);
```

```python
sns.scatterplot(x='carat', y='price', hue='cut',
                data=diamonds
                );
```

```python
sns.scatterplot(x='carat', y='price', hue='cut',
                data=diamonds,
                palette=['purple', '#55CCCC']
                );
```

```python
sns.scatterplot(x='carat', y='price', hue='depth',
                data=diamonds
                );
```

```python
sns.scatterplot(x='carat', y='price',
                data=diamonds,
                s=100
                );
```

```python
sns.scatterplot(x='carat', y='price', style='cut',
                data=diamonds,
                s=100
                );
```

```python
sns.scatterplot(x='carat', y='price', size='cut',
                data=diamonds,
                s=100
                );
```

```python
sns.scatterplot(x='carat', y='price', size='cut',
                data=diamonds,
                sizes=[150, 50]
                );
```

```python
sns.scatterplot(x='carat', y='price',
                data=diamonds,
                s=200,
                marker='*',
                edgecolor='black'
                );
```
