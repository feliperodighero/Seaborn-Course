```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.shape
```

```python
cars.head()
```

# BoxPlot

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4, 6, 8])]
```

```python
sns.boxplot(x=cars.mpg); #25% - 50% - 75%
```

```python
cars.mpg.describe() #25% - 50% - 75%
```

```python
sns.boxplot(x=cars.origin, y=cars.mpg);
```

```python
sns.boxplot(x='origin', y='mpg', data=cars);
```

```python
sns.boxplot(x='origin', y='mpg', hue='cylinders', data=cars);
```

```python
cars.model_year.describe()
```

```python
cars['newer_model'] = cars.model_year > 76
```

```python
sns.boxplot(x='origin', y='mpg', hue='newer_model', data=cars);
```

# Styling

```python
sns.boxplot(x='mpg', y='origin',
            data=cars
            );
```

```python
sns.boxplot(x='mpg', y='origin',
            data=cars,
            order=['japan', 'europe', 'usa']
            );
```

```python
sns.boxplot(x='mpg', y='origin', hue='newer_model',
            data=cars,
            order=['japan', 'europe', 'usa'],
            hue_order=[True, False]
            );
```

```python
sns.boxplot(x='mpg', y='origin',
            data=cars,
            color='g'
            );
```

```python
sns.boxplot(x='mpg', y='origin', hue='newer_model',
            data=cars,
            color='g'
            );
```

```python
sns.boxplot(x='mpg', y='origin',
            data=cars,
            width=0.5
            );
```

```python
sns.boxplot(x='mpg', y='origin',
            data=cars,
            linewidth=2.5
            );
```

```python
sns.boxplot(x='mpg', y='origin',
            data=cars,
            whis=1
            );
```

```python
sns.boxplot(x='mpg', y='origin',
            data=cars,
            fliersize=2
            );
```

```python
sns.boxplot(x='mpg', y='origin',
            data=cars,
            showcaps=False
            );
```
